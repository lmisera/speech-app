```python
def extract_data_from_pdf(pdfs):
    entry_pattern = re.compile(
    r"^(?=.*\b2020\b)(?=.*\b(Canton|Copley|Highland|Northwest|Perry|Stow|Tusky|Wadsworth|Wooster)\b).*?Judge:(.+?)\nEntry: (\d+)(.+?)\n(.*?)(?=Canton|Copley|Highland|Northwest|Perry|Stow|Tusky|Wadsworth|Wooster|Entry:\d+|$)",
    re.DOTALL | re.MULTILINE
    )
    
    print(entry_pattern)
    
    all_text = ""
    
    for i in pdfs:
        doc = fitz.open(i)
        for page in doc:
            all_text+=page.get_text()
        doc.close()
    
    return(entry_pattern.findall(all_text))

data = extract_data_from_pdf([f'{directory}/Highland-Speech-Ballots.pdf'])

for i in data[0]:
    print(i+"\n\n")
```
